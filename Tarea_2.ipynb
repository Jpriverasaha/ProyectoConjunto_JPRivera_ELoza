5. Use filter to determine the percentage of Fahrenheit temperatures in a list are within the range 32 to 80


```python
import numpy as np
np.random.seed(55555)
temperatures = np.random.uniform(25,110,30)
print(temperatures)
```

```python
## Juan Pablo
r=list(filter(lambda x: (32< x < 80) , temperatures))
porcentaje=(len(r)/len(list(temperatures)))

r, porcentaje*100
```

```python
# 5 Eduardo

lista= filter(lambda x:x>=32 and x<=80,temperatures)
print('El porcentaje de temperaturas entre 32 y 80 es:%.2f porciento' %((len(list(lista))/len(temperatures))*100))
```
